```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
%matplotlib inline

import scipy
import sklearn as sks
import statsmodels
import datetime
```

```python
bougainvilleUsage=pd.read_clipboard()
```

```python
bougainvilleUsage.head()
```

```python
sns.jointplot(x='CDD65',y='Usage',data=bougainvilleUsage,kind="reg")
```

```python
sns.jointplot(x='CDD75',y='Usage',data=bougainvilleUsage,kind="reg")
```

```python
bougainvilleUsage.corr(method='pearson')
```

```python
bougainvilleUsage['DateTime']=pd.to_datetime(bougainvilleUsage['Month'])
```

```python
bougainvilleUsage['DaysInMonth']=bougainvilleUsage['DateTime'].dt.daysinmonth
```

```python
import statsmodels.formula.api as smf
```

```python
model65=smf.ols(formula="Usage~DaysInMonth+CDD65",data=bougainvilleUsage)

model75=smf.ols(formula="Usage~DaysInMonth+CDD74",data=bougainvilleUsage)

model65fit=model65.fit()

model75fit=model75.fit()
```

```python
model65fit.summary()
```

```python
model75fit.summary()
```

```python
plt.plot(bougainvilleUsage['CDD65'],bougainvilleUsage['Usage'],'rs',label='Actual Usage')
plt.plot(bougainvilleUsage['CDD65'],model65fit.predict(bougainvilleUsage),'bo',label='Forecast')
plt.legend(loc=4)
```

```python
plt.plot(bougainvilleUsage['CDD75'],bougainvilleUsage['Usage'],'rs',label='Actual Usage')
plt.plot(bougainvilleUsage['CDD75'],model75fit.predict(bougainvilleUsage),'bo',label='Forecast')
plt.legend(loc=4)
```

```python
pd.DataFrame(model75fit.predict(bougainvilleUsage))
```
